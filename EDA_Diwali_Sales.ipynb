### Data Cleaning

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.drop(['Status','unnamed1'], axis = 1,inplace = True)
```

```python
df.isnull()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace= True)
```

### EDA

```python
df.columns
```

```python
sns.set( rc = {'figure.figsize':(4,5)})
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount')
```

```python
sns.set( rc = {'figure.figsize':(4,5)})
sales_gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount')

sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

#### It show that females spent more than twice than males

### Age

```python
sns.countplot(data = df, x = 'Age Group')
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
a = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

b = sns.barplot(x= 'Age Group', y = 'Orders', data = sales_age)

for bars in b.containers:
    b.bar_label(bars)
```

### State

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(7)

sns.set(rc= {'figure.figsize':(16,5)})
c = sns.barplot(data = sales_state, x ='State', y = 'Orders')

# this code below is used to get numbers on the bars.
for bars in c.containers:
    c.bar_label(bars)
```

### Amount

```python
sales_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set( rc = {'figure.figsize':(16,5)})
d = sns.barplot(data = sales_amount,x ='State', y = 'Amount' )

for bars in d.containers:
    d.bar_label(bars)
```

```python
df.columns
```

#### Here are top 10 states according to amount spent and on basis of number of orders

### Marital Status

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values( by = 'Amount', ascending = False)

sns.set( rc = {'figure.figsize':(6,5)})

e = sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount' , hue = 'Gender')

for bars in e.containers:
    e.bar_label(bars)
```

#### Here our analysis shows married women spent more money

### Occupation

```python
sns.set(rc = {'figure.figsize':(20,8)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_work = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(['Amount'], ascending = False)

ax = sns.barplot(data = sales_work, x = 'Occupation', y = 'Amount')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Products Category 

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(28,15)})
sns.countplot(data = df, x = 'Product_Category')
```

```python
sales_prod = df.groupby(['Product_Category'], as_index = False).sum('Amount').sort_values(by = 'Amount', ascending = False).head(10)
ax = sns.barplot(data = sales_prod, x = 'Product_Category', y = 'Amount')

sns.set(rc = {'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Most sales is for food products.

#### Conclusion:
##### Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
